# **[BA1K:](https://rosalind.info/problems/ba1k/)** Generate the Frequency Array of a String


```python
t = ['A','C','G','T']

# for item in s:
#   print(''.join(item))
def search_space(text,pattern,k):
  r =0
  for i in range(len(text)-k+1):
    s = text[i:i+k]
    if s==pattern:
      r+=1
  return r

def freq_arr(text,k):
  s = list(product(t,repeat=k))
  dict1 = {}
  for item in s:
    str1 = ''.join((item))
    c = search_space(text,str1,k)
    dict1[str1]=c

  arr = []
  for item,val in dict1.items():
    arr.append(val)
  return arr

# with open("output.txt","w") as f:
#   for item in freq_arr('CCTTAACGCCGAGGGCCTAAAGCCCTCCACCCTCTGGTGGGTGACGGACTTCCAAACAGGGATATAGTTACCAGGCCGTGTCGAGACTGTCCAAAAAGCTTCATTTCATTTGTTTACGTATTTTGAGCTCGCAGACAGTCACACGCATGGAGCAAACAGGGGCAAAACGGGGCCCTCCGGGTCGTGCCGCTTAGTGAGGGTGTTTCCAGACTCCGCGTGAATAACCATGGCAAGTCCAGGCCACAGTAGTTCAACGATTGTGAGCATGTGGCCTACGGCTCTGTCCCCAGATCGGATAGTGAGATCCTGCTGGTCGACTCGTCGTCCTACCTCCGGAAGTAGGGCGCCTGATGCGCACTAAGGCAGACCCTGAGATACTAATGCCAGAGGCTGCTAATATTATCGTGCAATAACCTGGCGTTCGAGAAGAACGGTGCATACCGTTATTGAAGTTTAGCGCATGATTAAGTTGTTCCGAGGTTTGACCTGTCTGACTAGAGTTTCGTACCGACACTTCTTATTTCTGTAGAAGCAGTCGATACAGATCAACGCCTACGTCTCACTATCCGAACCTGGCCCATATAAGATGAGGCCCATTTTCGCAACCGCCGGCGCAGAGCCCGCGGGCCACAGTAACG',7):
#     f.write(str(item))
#     f.write('\n')
print(freq_arr('CCTTAACGCCGAGGGCCTAAAGCCCTCCACCCTCTGGTGGGTGACGGACTTCCAAACAGGGATATAGTTACCAGGCCGTGTCGAGACTGTCCAAAAAGCTTCATTTCATTTGTTTACGTATTTTGAGCTCGCAGACAGTCACACGCATGGAGCAAACAGGGGCAAAACGGGGCCCTCCGGGTCGTGCCGCTTAGTGAGGGTGTTTCCAGACTCCGCGTGAATAACCATGGCAAGTCCAGGCCACAGTAGTTCAACGATTGTGAGCATGTGGCCTACGGCTCTGTCCCCAGATCGGATAGTGAGATCCTGCTGGTCGACTCGTCGTCCTACCTCCGGAAGTAGGGCGCCTGATGCGCACTAAGGCAGACCCTGAGATACTAATGCCAGAGGCTGCTAATATTATCGTGCAATAACCTGGCGTTCGAGAAGAACGGTGCATACCGTTATTGAAGTTTAGCGCATGATTAAGTTGTTCCGAGGTTTGACCTGTCTGACTAGAGTTTCGTACCGACACTTCTTATTTCTGTAGAAGCAGTCGATACAGATCAACGCCTACGTCTCACTATCCGAACCTGGCCCATATAAGATGAGGCCCATTTTCGCAACCGCCGGCGCAGAGCCCGCGGGCCACAGTAACG',7))
```

# **[BA1L:](https://rosalind.info/problems/ba1l/)** Implement PatternToNumber

```python
t = {'A':0,'C':1,'G':2,'T':3}

def digit_find(pattern):
  number = 0
  for i in pattern:
    digit = t[i]
    number = number*4+digit
  return number

digit_find('CGACTATAGGGACAACCCTGAAC')
```

# **[BA1M:](https://rosalind.info/problems/ba1m/)** Implement NumberToPattern


```python
t ={
    0:'A',
    1:'C',
    2:'G',
    3:'T'
}

def find_pattern(index,k):
  arr = []
  for i in range(k):
    s = index%4
    arr.append(t[s])
    index//=4

  return ''.join(reversed(arr))

find_pattern(7529,11)
```

# **[BA1N:](https://rosalind.info/problems/ba1n/)** Generate the d-Neighborhood of a String

```python
def ham_dis(str1,str2):
  c =0
  for i in range(len(str1)):
    if(str1[i]!=str2[i]):
      c+=1
  return c

def d_neighbors(text,d):
  t = {'A':0,'C':1,'G':2,'T':3}
  arr = []
  patterns = list(product(t,repeat=len(text)))
  for item in patterns:
    str1 = ''.join(item)
    x = ham_dis(str1,text)
    if x<=d:
      arr.append(str1)
  return arr

for item in d_neighbors('GAGAACGC',3):
  print(item)
```
