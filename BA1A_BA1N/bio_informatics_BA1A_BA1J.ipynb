# **[BA1H:](https://rosalind.info/problems/ba1h/)** Find All Approximate Occurrences of a Pattern in a String

```python
def find_occ(pattern,text,d):
  li = []
  for i in range(len(text)-len(pattern)+1):
    s = text[i:(i+len(pattern))]
    r = dis(pattern,s)
    if r<=d:
      li.append(i)

  return li

x = 'ATTCTGGA'
y = 'CGCCCGAATCCAGAACGCATTCCCATATTTCGGGACCACTGGCCTCCACGGTACGGACGTCAATCAAATGCCTAGCGGCTTGTGGTTTCTCCTACGCTCC'

find_occ(x,y,3)

# with open('input.txt',"w") as f:
#   for item in find_occ(x,y,4):
#     f.write(str(item))
#     f.write('\n')

```
